```python
from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments, Trainer
from datasets import load_dataset
import torch
```

모델, 토크나이저

```python
hugging_face_token = "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"

model_name = "meta-llama/Llama-3.1-8B"
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.1-8B")
model2 = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-3.1-8B")
```

GPU 설정

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model2.to(device)
```

데이터 로드 및 전처리

```python
recipe_dataset = load_dataset("csv", data_files="/content/drive/MyDrive/NewRecipe/full_dataset.csv")
```

```python
# 전처리 함수
def preprocess_function(example):
    # 입력 텍스트는 'ner' 컬럼
    input_text = example['NER']
    # 출력 텍스트는 'title'과 'direction'을 합친 것
    output_text = f"{example['title']} : {example['directions']}"

    # 토큰화 및 포맷
    inputs = tokenizer(input_text, padding="max_length", truncation=True, max_length=128)
    labels = tokenizer(output_text, padding="max_length", truncation=True, max_length=256)

    # 라벨 설정
    inputs["labels"] = labels["input_ids"]
    return inputs

# def tokenize_function(example):
#     return tokenizer(example["text"],
#                      truncation=True,
#                      padding="max_length",
#                      max_length=512)

# tokenized_datasets = recipe_dataset.map(tokenize_function, batched=True)
```

```python
# 패딩 토큰 추가
tokenizer.add_special_tokens({'pad_token': '[PAD]'})

# 데이터셋에 대한 토큰화 적용
tokenized_datasets = recipe_dataset.map(preprocess_function,
                                        batched=True,
                                        batch_size=256)
```

# 학습 파라미터 설정

```python
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=3,
    weight_decay=0.01,
    push_to_hub=False,
)

# Trainer 설정
trainer = Trainer(
    model=model2,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
)
```

학습

```python
trainer.train()
```
